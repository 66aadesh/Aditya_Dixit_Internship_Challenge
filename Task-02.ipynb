# Splitting the dataframe

```python
# Independent Varible dataset
X = train_data.iloc[:, :-1].values
X
```

```python
# Dependent Variable dataset
y = train_data.iloc[:, -1].values
y
```

```python
# Train/Test = 80/20

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# K Nearest Neighbour implementation on train.xlsx

```python
# K Nearest Neighbour from the Scikit Learn Library

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

- K Nearest Neighbours - Predicting the labels/classification using X_test

```python
y_pred = classifier.predict(X_test)
```

```python
# Comparison: - y_pred vs y_test

compare = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)

pred_vs_test = pd.DataFrame(compare)

pred_vs_test
```

- Confusion matrix & Accuracy - K Nearest Neighbours

```python
# K Nearest Neighbours - Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print(cm.shape)
print(cm)
```

```python
# K Nearest Neighbours - Accuracy

knn_accuracy = (accuracy_score(y_test, y_pred))*100

print(knn_accuracy)
```

# K Nearest Neighbour implementation on test.xlsx
